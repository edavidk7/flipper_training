```python
import torch
import OpenEXR
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
dest = Path("../heightmap_files")
```

```python
p = "/Users/davidkorcak/Documents/ctu/bachelors/blender/textures/ant01_heightmap.exr"
i = OpenEXR.InputFile(p)
hm = torch.frombuffer(i.channel("R"), dtype=torch.float32) * 0.8
hm = hm.reshape(
    i.header()["dataWindow"].max.y - i.header()["dataWindow"].min.y + 1, i.header()["dataWindow"].max.x - i.header()["dataWindow"].min.x + 1
)
torch.save(hm, dest / "natural_simple.pt")
plt.imshow(hm, cmap="gray")
plt.show()
```
